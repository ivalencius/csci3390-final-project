# Plots

```python
import matplotlib.pyplot as plt
import scienceplots
import numpy as np
plt.style.use(["nature", "notebook"])
```

```python
graph_sizes = np.array([
    2671,
    35324,
    22301964,
    42851237,
    63555749,
    117185083
])
rounds = np.array([
    8,
    14,
    26,
    38,
    np.nan,
    43
])
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(graph_sizes, rounds, marker="o")
ax.plot(graph_sizes, np.log(graph_sizes), linestyle=":", color="gray", label="log(n)")
ax.plot(graph_sizes, 2.5*np.log(graph_sizes), linestyle="--", color="gray", label="3log(n)")

ax.set_xscale("log")
# ax.set_yscale("log")
ax.set_xlabel("Graph size")
ax.set_ylabel("Rounds")
ax.legend()
ax.set_title("Rounds vs. Graph size")
plt.show()
```
